```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmod(x):
  return 1/(1+np.exp(-x))
x=np.linspace(-10,10,100)
plt.plot(x,sigmod(x))
plt.axis('tight')
plt.title('activaton function: sigmoid')
plt.show()
```

```python
def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,10)
plt.plot(x, tanh(x))
plt.title("Activation functions: Tanh(x)")
plt.show()
```

```python
def ReLU(x):
  s=[]
  for i in x:
    if i<0:
      s.append(0)
    else:
      s.append(1)
  return s
x=np.linspace(-40,80,100)
print(x)
print()
plt.plot(x, ReLU(x))
plt.title("Activation Function: ReLU")
plt.show()
```

```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x), axis=0)
x=np.linspace(-20,80,100)
print(softmax(x))
plt.plot(x, softmax(x))
plt.title("Activation Function: softmax")
plt.show()
```
